```python
market_cap_75th = 23549.90
market_cap_median = 9885.05
market_cap_25th = 4843.575

sales_qtr_75th = 2730.195
sales_qtr_median = 1137.17
sales_qtr_25th = 534.91

# Function to categorize market cap
def categorize_market_cap(value):
    if value >= market_cap_75th:
        return "Large Cap"
    elif value >= market_cap_median:
        return "Mid Cap"
    elif value >= market_cap_25th:
        return "Small Cap"
    else:
        return "Very Small Cap"

# Function to categorize sales quarter
def categorize_sales_qtr(value):
    if value >= sales_qtr_75th:
        return "High Sales"
    elif value >= sales_qtr_median:
        return "Medium Sales"
    elif value >= sales_qtr_25th:
        return "Low Sales"
    else:
        return "Very Low Sales"

# Apply the functions to create new columns
df['Market_Cap_Category'] = df['Mar Cap - Crore'].apply(categorize_market_cap)
df['Sales_Qrt_Category'] = df['Sales Qtr - Crore'].apply(categorize_sales_qtr)

# Display the updated DataFrame
df.head()
```
